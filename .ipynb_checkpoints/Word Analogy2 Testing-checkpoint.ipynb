```python
def euclidean_dist(vector_1, vector_2):
#     vector_1 = np.array(vector_1)
#     vector_2 = np.array(vector_2)
    return np.sqrt(np.sum((vector_1 - vector_2)**2))

def manhattan_dist(vector_1, vector_2):
#     vector_1 = np.array(vector_1)
#     vector_2 = np.array(vector_2)
    return np.sum(np.abs(vector_2 - vector_1))

def cosine_dist(vector_1, vector_2):
#     vector_1 = np.array(vector_1)
#     vector_2 = np.array(vector_2)
    return np.dot(vector_1, vector_2)
```

```python
def compare_and_return_fourth_vector(word1,word2,word3, similarity_type):    
    # also, lower case all words before checking!
    word1 = word1.lower()
    word2 = word2.lower()
    word3 = word3.lower()
    # need to have edge case for: if the word's vector exists or not?
    if (word1 not in df.columns) or (word2 not in df.columns) or (word3 not in df.columns):
        return ''
    
    # convert into vector first
    
    # big problem here ! with numpy arrays or pandas columns in another format
    v1=df[word1]
    v2=df[word2]
    v3=df[word3]
#     print(v1)
#     print(v2)
#     print(v3)
    predicted_vect = v3 + v2 - v1 # gettting predicted 4th vector
    print(predicted_vect)
    return 'ok break'
    if similarity_type == 0:
        '''
            euclidean distance comparison
        '''
        initial_dist = euclidean_dist(vector_1=predicted_vect, vector_2=np.array(df[df.columns[0]])) # initilise
        word = df.columns[0]
        for i in df: 
            print(i)
            #if i.lower() != word3: # so that it won't predict 3rd word haha!
            temp_euclidean_dist = euclidean_dist(vector_1=predicted_vect,vector_2=np.array(df[i]))
            if temp_euclidean_dist < initial_dist:
                word = i
                initial_dist = temp_euclidean_dist
                print(i)
        return word
    
    if similarity_type == 1:
        '''
            manhattan distance comparison
        '''
        initial_dist = manhattan_dist(vector_1=predicted_vect, vector_2=df[df.columns[0]]) # initilise
        word = df.columns[0]
        for i in df:
            if i != word3: # so that it won't predict 3rd word haha!
                temp_manhattan_dist = manhattan_dist(vector_1=predicted_vect,vector_2=df[i])
                if temp_manhattan_dist < initial_dist:
                    word = i
                    initial_dist = temp_manhattan_dist
        return word
    
    if similarity_type == 2:
        '''
            cosine distance comparison
        '''
        initial_dist = cosine_dist(vector_1=predicted_vect, vector_2=df[df.columns[0]]) # initilise
        word = df.columns[0]
        for i in df:
            if i != word3: # so that it won't predict 3rd word haha!
                temp_cosine_dist = cosine_dist(vector_1=predicted_vect,vector_2=df[i])
                if temp_cosine_dist < initial_dist:
                    word = i
                    initial_dist = temp_cosine_dist
        return word
    
    if similarity_type == 3:
        '''
            average of all distances comparison
        '''
        initial_avg_dist = (euclidean_dist(vector_1=predicted_vect, vector_2=df[df.columns[0]]) + manhattan_dist(vector_1=predicted_vect, vector_2=df[df.columns[0]]) + cosine_dist(vector_1=predicted_vect, vector_2=df[df.columns[0]]))/3 # initilise
        word = df.columns[0]
        for i in df:
            print(i)
            if i != word3: # so that it won't predict 3rd word haha!
                temp_avg_dist = (euclidean_dist(vector_1=predicted_vect, vector_2=df[i]) + manhattan_dist(vector_1=predicted_vect, vector_2=df[i]) + cosine_dist(vector_1=predicted_vect, vector_2=df[i]))/3
                if temp_avg_dist < initial_avg_dist:
                    word = i
                    initial_avg_dist = temp_avg_dist
        return word
    
'''
    thinks to improve upon:
        - make a distance class and use it that way OOP concept
        - edge cases for compariing 
        - dynamic programming to do distance calcualtion of each vector with that predicted one and
            - find min distance from there!
'''
```
